```python
#To see just the numbers of missing content
df.isnull().sum()
```

```python
#Checking for any missing values via a diagram
df.isnull().sum().plot(kind="bar", figsize=(8,4))
# if we have missing data in column it will be shown in the bar chart, if there is no missing data, the bar chart will be empty.
```

```python
#Data Cleaning
#Dropping all the missing values in rows
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#Checking duplicated values
df.duplicated().sum()
```

```python
#another way to do it.
df[df.duplicated()]
```

```python
#Deleting all duplicated records 
df.drop_duplicates(inplace=True)
df.duplicated().sum()
#df[df.duplicated()] Would return 0 rows because all duplicates have been removed
```

```python
#Now to check it again with the help of a diagram
df.isnull().sum().plot(kind="bar", figsize=(8,4))
```

```python
df.dtypes
```

```python
#Changing data type of id, host id to object, ie type casting here
df["id"] = df["id"].astype(object)

df["host_id"] = df["host_id"].astype(object)
```

```python
df.dtypes #New ones
```

```python
#Identifying the outliers first

df =df[df["price"] < 1500]

sns.boxplot(data=df, x="price")
```

```python
#Data Analysis
#For price distribution
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='price', kde=True, bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show
```
