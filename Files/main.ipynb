```python
#Checking for any missing values via a diagram
df.isnull().sum().plot(kind="bar", figsize=(8,4))
# if we have missing data in column it will be shown in the bar chart, if there is no missing data, the bar chart will be empty.
```

```python
#Data Cleaning
#Dropping all the missing values in rows
df.dropna(inplace=True)
```
